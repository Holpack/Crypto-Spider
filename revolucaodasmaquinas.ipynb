```python
# prompt: print a plot with random numbers

import matplotlib.pyplot as plt
import random

# Generate random data
x = range(10)  # Example x-values
y = [random.randint(1, 100) for _ in range(10)] # Example y-values (random integers between 1 and 100)


# Create the plot
plt.plot(x, y)


# Customize the plot (optional)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Random Number Plot")
plt.grid(True)

# Display the plot
plt.show()
```
